## What is matplotlib
- The matplotlib package helps you make visually appealing representations of the data you are working with

## Line graphs

```python
# Import matplotlib
import matplotlib.pyplot as plt
```

```python
# Making a line graph
x_values = [0,1,2,3,4,5]
squares = [0,1,4,9,16,25]
plt.plot(x_values, squares) # x_
plt.show()
```

## Making a scatter plot
- The scatter()function takes __a list of x values__ and __a list of y values__ and __a variety of optional arguments__
- The s=10 argument controls the size of each point

```python
x_values = list(range(100))
squares = [x**2 for x in x_values]

plt.scatter(x_values, squares, s=5)
plt.show()
```

## Customizing plots
- Plots can be customized in a wide variety of ways
- Any element of a plot can be customized

```python
# Adding titles and labels, and scaling axes
x_values = list(range(100))
squares = [x**2 for x in x_values]
plt.scatter(x_values, squares, s=5)

plt.title("Square Numbers", fontsize=24) # title
plt.xlabel("Value", fontsize=18) # x axis label
plt.ylabel("Square of Value", fontsize=18) # y axis label
plt.tick_params(axis='both', which='major', labelsize=14) # axis label
plt.axis([0,140,0,11000]) # range of x-axis and y-axis
plt.show()
```

### Using a colormap
- A colormap varies the point colors from one shade to another, based on a certain value for each points
- The value used to determine the color of each point is passed to the c argument, and __the cmap argument specifies which colormap to user__
- The __edgecolor ='none' argument__ removes the black outline from each points

```python
plt.scatter(x_values, squares, c=x_values,
           cmap=plt.cm.Greens, edgecolor='none',s=10)
plt.show()
```

## Emphasizing points
- You can plot as much data as you want on one plot. Here we replot the first and last points larger to emphasize them

```python
x_values = list(range(100))
squares = [x**2 for x in x_values]
plt.scatter(x_values, squares, c=squares,
           cmap=plt.cm.Blues, edgecolor='none', s = 10)

# the first point is emphasized
plt.scatter(x_values[0], squares[0], c='orange', edgecolor='none', s=100)
# the last point is emphasized
plt.scatter(x_values[-1], squares[-1], c='red', edgecolor='none', s=100)

plt.title("Square Numbers", fontsize=24)
plt.show()
```

### Removing axes
- You can customize or remove axes entirely. Here's how to access each axis and hide it

```python
x_values = list(range(100))
squares = [x**2 for x in x_values]
plt.scatter(x_values, squares, c=squares,
           cmap=plt.cm.Blues, edgecolor='none', s = 10)

# the first point is emphasized
plt.scatter(x_values[0], squares[0], c='orange', edgecolor='none', s=100)
# the last point is emphasized
plt.scatter(x_values[-1], squares[-1], c='red', edgecolor='none', s=100)

plt.title("Square Numbers", fontsize=24)

plt.axes().get_xaxis().set_visible(False) # how to make your x-axis disappear
plt.axes().get_yaxis().set_visible(False) # how to make your y-axis disappear
plt.show()
```

- To do so, replace __plt.show()__ with __plt.savefig()__
- The bbox_inches ='tight' argument trims extra whitespace from the plot

```python
# Adding titles and labels, and scaling axes
x_values = list(range(100))
squares = [x**2 for x in x_values]
plt.scatter(x_values, squares, s=5)

plt.title("Square Numbers", fontsize=24) # title
plt.xlabel("Value", fontsize=18) # x axis label
plt.ylabel("Square of Value", fontsize=18) # y axis label
plt.tick_params(axis='both', which='major', labelsize=14) # axis label
plt.axis([0,140,0,11000]) # range of x-axis and y-axis

plt.savefig('square.png', bbox_inces='tight')
```

## Mutiple plots
- You can make as many plots as you want on one figure
- When you maake multiple plots, you can emphasize relationships in the data
- For example you can fill the space between two sets of data

```python
# plotting two sets of data
x_values = list(range(100))
squares = [x**2 for x in x_values]
cubes =[x**3 for x in x_values]

plt.scatter(x_values, squares ,c='blue',
           edgecolor='none', s=25)
plt.scatter(x_values, cubes, c='red',
           edgecolor='none', s = 25)

plt.title("Square and Cube Numbers", fontsize=24)
plt.ylabel("Square of Value/Cube of Value", fontsize=18) # y axis label


plt.axis([0,11,0,1100])
plt.show()
```

### Fill the space between datasets
- The fill_between() method fills the space between two data sets
- It takes a series a x-values and two series of y-values
- It takes a facecolor to use for the fill
- An optional alpha argument that controls the color's transparency

```python
# plotting two sets of data
x_values = list(range(100))
squares = [x**2 for x in x_values]
cubes =[x**3 for x in x_values]

plt.scatter(x_values, squares ,c='blue',
           edgecolor='none', s=25)
plt.scatter(x_values, cubes, c='red',
           edgecolor='none', s = 25)

plt.title("Square and Cube Numbers", fontsize=24)
plt.ylabel("Square of Value/Cube of Value", fontsize=18) # y axis label
plt.xlabel("Value", fontsize=18) # y axis label


plt.axis([0,11,0,1100])

plt.fill_between(x_values, cubes, squares, facecolor='yellow', alpha = 0.10)

plt.show()
```

## Working with dates and times
- Many interesting datasets have a date or time as the x-value.
- Python's datetime module helps you work with this kind of data

### Generating the current date
- The __datetime.now()__ function returns a datetime object representing the current date and time

```python
# import datetime
from datetime import datetime as dt
```

```python
today = dt.now()
date_string = dt.strftime(today, '%x/%d/%Y')
print(date_string)
```

### Generating a specific data
- You can also generate a datetime object for any date and time you want 
- The positional order of arguments is year, month, and the day. 
- The hour, minute, second and minisecond arguments are optional

```python
new_years = dt(2010, 9 , 1)
print(new_years)
```

```python
fall_equinox = dt(year=2003, month=9, day=1)
print(fall_equinox)
```

## Plotting high temperature
- The following code creates a list of dates and a corresponding list of high temperature. It then plots the high temperature, with the data labels displayed in a specific format

```python
from matplotlib import dates as mdates
```

```python
dates = [
    dt(2016, 6, 21), dt(2016, 6, 22),
    dt(2016,6,23), dt(2016, 6, 24),
]

highs = [57,68,64,59]

fig = plt.figure(dpi=128, figsize=(10,6))
plt.plot(dates, highs, c='red')
plt.title("Daily High Temps", fontsize=24)
plt.ylabel("Temp (F)", fontsize =16)

x_axis = plt.axes().get_xaxis()
x_axis.set_major_formatter(
        mdates.DateFormatter('%b %d %Y')
        )

fig.autofmt_xdate()

plt.show()
```

## Multiple plots in one figure
- You can include as many individual graphs in one figure as you want
- This is useful, especially when you are comparing related datasets

### Sharing an x-axis
- The __plt.subplots()__ function returns a figure object and a tuple of axes
- Each set of axes corresponds to a separate plot in the figure
- The __first argument control the number of rows and columns generated in the figure__

```python
# Multiple plots in one figure by sharing x
x_values = list(range(100))
squares = [x**2 for x in x_values]
cubes =[x**3 for x in x_values]

fig, axarr = plt.subplots(2,1, sharex=True) # first argument = how many corresponding to x-axis

axarr[0].scatter(x_values, squares)
axarr[0].set_title('Squares')

axarr[1].scatter(x_values, cubes, c='red')
axarr[1].set_title('Cubes')

plt.show()
```

### Sharing a y-axis
- To share a y-axis, we use the sharey=True argument

```python
# Multiple plots in one figure by sharing y
x_values = list(range(100))
squares = [x**2 for x in x_values]
cubes =[x**3 for x in x_values]

fig, axarr = plt.subplots(1,2, sharey=True) # 2nd argument = how many corresponding to the y-axis

axarr[0].scatter(x_values, squares)
axarr[0].set_title('Squares')

axarr[1].scatter(x_values, cubes, c='red')
axarr[1].set_title('Cubes')

plt.show()
```
